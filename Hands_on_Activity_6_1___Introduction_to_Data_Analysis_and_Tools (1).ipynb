*   Mean



```python
# Write a comment per statistical function
def mean(dataset):
  return sum(dataset) / len(dataset)

print(mean(salaries))
```

*  Median

```python
# Write a comment per statistical function
def median(dataset):
    n = len(dataset)
    sorted_dataset = sorted(dataset)
    m = n // 2
    if n % 2 == 0:
        return (sorted_dataset[m-1] + sorted_dataset[m]) / 2
    else:
        return sorted_dataset[m]
print(median(salaries))
```

* Mode

```python
# Write a comment per statistical function
def mode(dataset):
    freq = {}
    for num in dataset:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1
    max_count = max(freq.values())
    modes = [num for num, count in freq.items() if count == max_count]
    return modes if len(modes) > 1 else modes[0]
print(mode(salaries))
```

* Sample Variance

```python
# Write a comment per statistical function
def Variance(dataset):
  sum = 0
  ave = mean(dataset)
  n = len(dataset)
  for i in dataset:
    x = (i - ave) ** 2
    sum += x
  return sum / (n - 1)

print(Variance(salaries))
```

* Sample Standard Deviation

```python
from math import sqrt
# Write a comment per statistical function
def SD(dataset):
  print(sqrt(Variance(dataset)))
  return (Variance(dataset)) ** 0.5

print(SD(salaries))
```

Using the same data, calculate the following statistics using the functions in the statistics module.

*   Coefficient of variation interquartile range
*   Quartile coefficient of dispersion



* Range

```python
def range(dataset):
    if len(dataset) == 0:
        return None
    else:
        min_value = min(dataset)
        max_value = max(dataset)
        return max_value - min_value

print(range(salaries))
```

* Coefficient of variation interquartile range

```python
def cv_iqr(dataset):
    if len(dataset) < 4:
        return None
    else:
        dataset.sort()
        n = len(dataset)
        q1 = dataset[n//4]
        q3 = dataset[3*n//4]
        iqr = q3 - q1
        if iqr == 0:
            return None
        else:
            m = sum(dataset) / n
            variance = sum((x - m)**2 for x in dataset) / (n - 1)
            std_dev = variance**0.5
            cv = std_dev / iqr * 100
            return cv
print(cv_iqr(salaries))
```

* Quartile coefficient of dispersion

```python
def qc_disp(dataset):
    if len(dataset) < 4:
        return None
    else:
        dataset.sort()
        n = len(dataset)
        q1 = dataset[n//4]
        q3 = dataset[3*n//4]
        iqr = q3 - q1
        if iqr == 0:
            return None
        else:
            qc_disp = (q3 - q1) / (q1 + q3)
            return qc_disp
print(qc_disp(salaries))
```
